```python
"""Character Type Counter"""

def count_file_elements(filename):
    vowels = "aeiouAEIOU"
    counts = {
        "Vowels": 0,
        "Consonants": 0,
        "Digits": 0,
        "Special Characters": 0
    }

    try:
        with open(filename, 'r') as file:
            text = file.read()
            
            for char in text:
                if char.isalpha():
                    if char in vowels:
                        counts["Vowels"] += 1
                    else:
                        counts["Consonants"] += 1
                elif char.isdigit():
                    counts["Digits"] += 1
                elif not char.isspace(): 
                    counts["Special Characters"] += 1
        
        for category, value in counts.items():
            print(f"{category}: {value}")
            
    except FileNotFoundError:
        print("File not found.")

count_file_elements('sample.txt')
```

```python
"Character Type Counter"

def merge_files(file1, file2, output_file):

    with open(file1, 'r') as f1:
        content1 = f1.read()
        
    with open(file2, 'r') as f2:
        content2 = f2.read()
        
    # Write both to the output file with the separator
    with open(output_file, 'w') as out:
        out.write(content1)
        out.write("\n----- End of File 1 -----\n")
        out.write(content2)

merge_files('notes.txt', 'data.txt', 'merged.txt')
```

```python
"""Word Search & Line Finder"""

def find_word_in_file(filename, target_word):
    found = False
    target_word = target_word.lower()
    
    with open(filename, 'r') as file:
        for line_number, line in enumerate(file, 1):
            words = [w.strip('.,!?;:').lower() for w in line.split()]
            
            if target_word in words:
                print(f"Word '{target_word}' found on line {line_number}")
                found = True
                
    if not found:
        print(f"The word '{target_word}' was not found in the file.")

find_word_in_file('document.txt', 'python')
```

```python
"""Student Grade Reporter"""

import csv

def get_grade(mark):
    if mark >= 80:
        return 'A'
    elif mark >= 60:
        return 'B'
    elif mark >= 40:
        return 'C'
    else:
        return 'D'

def display_student_report(filename):
    print(f"{'Name':<15} {'Mark':<8} {'Grade':<5}")
    print("-" * 30)

    try:
        with open(filename, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                name = row['name']
                mark = int(row['mark'])
                grade = get_grade(mark)
                print(f"{name:<15} {mark:<8} {grade:<5}")
    except FileNotFoundError:
        print("Error: The file was not found.")

display_student_report('students.csv')
```

```python
"""Most Frequent Word Finder"""

def find_most_frequent_word(filename):
    word_counts = {}

    with open(filename, 'r') as file:
        for line in file:
            words = line.lower().split()
            
            for word in words:                
                if word:
                    word_counts[word] = word_counts.get(word, 0) + 1

    if not word_counts:
        print("The file is empty.")
        return

    # Find the word with the maximum value
    most_common = max(word_counts, key=word_counts.get)
    count = word_counts[most_common]

    print(f"Most frequent word: '{most_common}'")
    print(f"Occurrences: {count}")


find_most_frequent_word('speech.txt')
```

```python
"""Unique Word Extractor"""

def extract_unique_sorted_words(input_file, output_file):
    with open(input_file, 'r') as infile:
        # Read the whole file and split into a list of words
        content = infile.read()
        words = content.split()
        
        unique_words = set(word.lower() for word in words)
        
    # Sort the unique words alphabetically
    sorted_unique_words = sorted(list(unique_words))
    
    with open(output_file, 'w') as outfile:
        for word in sorted_unique_words:
            outfile.write(word + '\n')

extract_unique_sorted_words('input.txt', 'unique_words.txt')
```

```python
"""Vowel-Started Line Counter"""

def count_vowel_lines(filename):
    vowels = "AEIOU"
    count = 0
    
    with open(filename, 'r') as file:
        for line in file:
            # strip() removes leading whitespace
            clean_line = line.strip()
            
            # Check if the line is not empty and starts with a vowel
            if clean_line and clean_line[0].upper() in vowels:
                count += 1
                
    return count


total = count_vowel_lines('notes.txt')
print(f"Number of lines starting with a vowel: {total}")
```

```python
"""Line Reverser"""

def reverse_file_lines(input_filename):
    with open(input_filename, 'r') as infile:
        lines = infile.readlines()
    
    with open('reversed.txt', 'w') as outfile:
        for line in lines:
            # strip() removes the newline character so it doesn't end up at the start
            reversed_line = line.strip()[::-1]
            outfile.write(reversed_line + '\n')

reverse_file_lines('example.txt')
```

```python
"""Salary avg"""
import csv

def average_salary_per_department(file_name):
    dept_salaries = {}  # Stores lists of salaries: {dept: [sal1, sal2]}

    with open(file_name, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            dept = row['department']
            salary = float(row['salary'])
            
            if dept not in dept_salaries:
                dept_salaries[dept] = []
            dept_salaries[dept].append(salary)

    print("Average Salary per Department:")
    for dept, salaries in dept_salaries.items():
        avg = sum(salaries) / len(salaries)
        print(f"{dept}: {avg:.2f}")

average_salary_per_department('employees.csv')
```

```python
"""LIbrary management system"""

class Library:
    def __init__(self, issue, ret, search):
        self.issue = issue
        self.ret = ret
        self.search = search

    def issue_book(self):
        name = input("Enter the name of book : ")
        date = int(input("Enter the date of issue : "))
        author = input("Enter the name of the author : ")
        isbn_num = int(input("Enter the isbn number : "))

        try:
            with open("library_book.csv", "w") as f:
                info = f'{name},{date},{author},{isbn_num}'
                f.write(info)
        except:
            print("Error Try again")


class Library_two(Library):
    def search_book(self): 
        name = input("enter the book to search: ")
        try:
            f = open("library_books.csv", "r")
            content = f.read().lower()
            if name.lower() in content:
                print(f"Yes, '{name}' exists in library")
            else:
                print(f"{name} doesn't exist")
            f.close()
        except:
            print("Error")

    def ret_books(self):
        books_ret = input("Enter the books to return: ")
        try:
            f = open("taken_books.txt", "r")
            taken_book = f.readlines()
            if books_ret in taken_book:
                print("Book returned successfully")
            else:
                print("Your book can't be returned")
            f.close()
        except:
            print("error")


def main():
    lib = Library_two("issue", "ret", "search") 

    while True:
        print("1. Issue book")
        print("2. Search")
        print("3. Return")
        print("4. Exit")

        choice = input("enter any number 1-4 : ") 
        if choice == "1": 
            lib.issue_book()
            break  
        elif choice == "2":
            lib.search_book()
            break
        elif choice == "3":
            lib.ret_books()
            break
        elif choice=="4":
            break

main()
```

```python
"""Employee management system """
class Employee :
    def __init__(self,empid,name,address,contact_number,spouse_name,number_of_children,salary ):
        self.empd = empid
        self.name = name 
        self.address = address 
        self.contact_number = contact_number
        self.spouse_name = spouse_name
        self.number_of_child = number_of_children
        self.salary = salary


def add_employee ():
    name =input("Enter name :")
    empid = input("Enter employee empd : ")
    contact_number =int(input("Enter employee contact"))
    address = input("Enter the address :")
    spouse_name = input("enter name of spouse : ")
    number_of_children = int(input("Enter no of child: "))

   
    try : 
        file = open("employees.csv","a")
        line =f'{name},{empid},{contact_number},{address},{spouse_name},{number_of_children}\n'
        file.write(line)
        file.close()
        print("Employee added")
    except : print("error try again")



def show_employee ():
    try: 
        file = open("employees.csv","r")
        lines=file.readlines()
        file.close()
        

        if not lines:
            print("No employees found in the file.")
        

        for line in lines:
            fields = line.strip().split(",")
            print("Name:", fields[0])
            print("Employee ID:", fields[1])
            print("Contact:", fields[2])
            print("Address:", fields[3])
            print("Spouse Name:", fields[4])
            print("Number of Children:", fields[5])


    except FileNotFoundError:
        print("Error")



while True:
    print("1. Add Employee")
    print("2. View Employee")

    choice= input("enter 1 or 2 : ")
    if choice == "1":
        add_employee()
    elif choice=="2":
        show_employee()
```
