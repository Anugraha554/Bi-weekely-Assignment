```python
"""
    Checks if a given year is a leap year according to the Gregorian calendar."""

year = 2025
is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
days = 366 if is_leap else 365
weeks = days // 7
extra = days % 7

print(f"{year} has {weeks} weeks and {extra} day(s) ({days} days total)")
```

```python
"""
    Analyzes a string to count the total number of alphabetic letters and digits."""

text = input("Enter a string: ")

letters = sum(c.isalpha() for c in text)
digits = sum(c.isdigit() for c in text)

print("Letters: ",letters)
print("Digits: ",digits)
```

```python
"""
    Executes a game where the user has 5 attempts to guess a random number.
    Uses feedback (Too high/Low) and random module generation.
    """

import random
num =random.randint(1,100)
for i in range(5):
    ans= int(input("Enter a integer : "))
    if ans == num :
        print("You are correct !")
        break
    elif ans<num:
        print("Too low !")
        continue
    else:
        print("Too high !")
        continue

print("Game over")
```
