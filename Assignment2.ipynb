```python
"""String Character Counter"""

def tkstr():
    text= input("Enter a string")
    letters=0
    digits=0
    specials=0
    for char in text:
        if char.isalpha():
            letters += 1
            print ("letter",letters)
        elif char.isdigit():
            digits += 1
            print("digit",digits)
        else:
            print("special")

tkstr()
```

```python
"""Perfect Number Checker"""

def is_perfect_number(n):
    if n < 1:
        return False
    
    divisor_sum = 0
    for i in range(1, n):
        if n % i == 0:
            divisor_sum += i
            
    return divisor_sum == n

print(is_perfect_number(6))  
print(is_perfect_number(10))  
```

```python
"""Palindrome Checker"""

def is_palindrome(n):
    if n < 0:
        return False
    
    original = n
    reversed_num = 0
    
    while n > 0:
        digit = n % 10
        reversed_num = reversed_num * 10 + digit
        n //= 10
        
    return original == reversed_num

print(is_palindrome(121))
print(is_palindrome(123))
```

```python
"""Even Number Sorter"""

def get_sorted_evens(numbers):
    even_numbers = [x for x in numbers if x % 2 == 0]
    return sorted(even_numbers)

nums = [7, 2, 10, 8, 1, 4]
print(get_sorted_evens(nums))
```

```python
"""Advanced Calculator Functions"""

import math

def square_root(n):
    return math.sqrt(n)

def absolute_value(n):
    return abs(n)

def floor_division(a, b):
    return a // b

def percentage(part, total):
    return (part / total) * 100

def average(a, b):
    return (a + b) / 2

print(square_root(16))
print(absolute_value(-10))
print(floor_division(10, 3))
print(percentage(20, 50))   
print(average(10, 20))      
```

```python
"""Unique Integer Input Handler"""

numb = set()

for i in range(10):
    num = int(input(f"Enter integer {i+1}: "))
    numb.add(num)

sorted_list = sorted(list(numb))

print("Sorted unique numbers:", sorted_list)
```

```python
"""Dictionary Merging"""

def merge_student_scores(student_names, student_scores):
    merged = {}
    for student_id in student_names:
        name = student_names[student_id]
        score = student_scores[student_id]
        merged[name] = score
    return merged

student_names = {1: 'Anish', 2: 'Priya', 3: 'Sujan'}
student_scores = {1: 88, 2: 92, 3: 79}

print(merge_student_scores(student_names, student_scores))
```

```python
"""weekly Sales Updates"""

def update_weekly_sales(sales_data, day, amount):
    if day in sales_data:
        sales_data[day] += amount
    else:
        sales_data[day] = amount
    return sales_data

weekly_sales = {"Monday": 100, "Tuesday": 150}

# Updating an existing day
update_weekly_sales(weekly_sales, "Monday", 50) 
# Adding a new day
update_weekly_sales(weekly_sales, "Wednesday", 200)

print(weekly_sales)
```

```python
"""dictionary operation"""
a = {'x': 5, 'y': 10}
b = {'z': 15, 'w': 20}
c = {'p': 25, 'q': 30}

# (a) Merge all dictionaries into a single one
combined = {**a, **b, **c}

# (b) Add a new pair r: 35
combined['r'] = 35

# (c) Double all the values in the resulting dictionary
for key in combined:
    combined[key] = combined[key] * 2

# (d) Delete the key y
if 'y' in combined:
    del combined['y']

# (e) Compute and print the average value of all remaining items
values = combined.values()
average = sum(values) / len(values)

print("Final Dictionary:", combined)
print("Average Value:", average)
```

```python
"""Set operation"""

setA = {1, 2, 3, 4, 5}
setB = {4, 5, 6, 7, 8}

# (a) Print the union and its size
union_set = setA | setB
print("Union:", union_set)
print("Union Size:", len(union_set))

# (b) Print the difference of setA - setB
print("Difference (setA - setB):", setA - setB)

# (c) Print the intersection of both sets
print("Intersection:", setA & setB)

# (d) Check whether setA is a subset of setB
print("Is setA a subset of setB?:", setA.issubset(setB))

# (e) Remove element 7 from setB and display the result
setB.discard(7)
print("Updated setB after removing 7:", setB)
```

```python
"""Common letter finder"""

def common_letters(word1, word2):
    # Convert both words to sets to find unique letters
    set1 = set(word1)
    set2 = set(word2)
    
    # Find the intersection (letters in both)
    common = set1 & set2
    
    # Return as a sorted list
    return sorted(list(common))

print(common_letters("apple", "orange"))
```
